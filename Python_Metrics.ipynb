Определить длину наибольшей последовательности одинаковых символов в строке

str = 'fffggggrrrrrrrrrtttfffff'; Ответ: 9

```python
def longest_sequence(s):
    if not s:
        return 0 #проверка на пустую строку
    
    max_length = 1 # инициализируем переменные 
    current_length = 1
    
    for i in range(1, len(s)):
        if s[i] == s[i-1]:  #cравниваем текущий символ s[i] с предыдущим символом s[i-1]
            current_length += 1 #счетчик, если символы одинаковые
        else: #если символы разные
            max_length = max(max_length, current_length) #сравниваем текущий максимум с текущей последовательностью 
            current_length = 1    #сохраняем бОльшее значение в max_length и начинаем считать новую последовательность 
    

    max_length = max(max_length, current_length) #проверяем, не является ли последняя последовательность самой длинной
    return max_length #Финальная проверка нужна, потому что цикл никогда не заходит в блок else для последней последовательности!


s = 'fffggggrrrrrrrrrtttfffff'
print(longest_sequence(s)) 
```

## Метрики ##

Есть продукт “почта”. Самая обычная почта, позволяющая отправлять и получать email сообщения. 
Есть desktop web, mobile web, android и ios приложение, также можно получать и отправлять письма из сторонних почтовых клиентов. 
В собственных клиентах показывается реклама, также есть возможность за небольшую ежемесячную плату отключить эту рекламу. 
Так получилось, что в продукте нет никакой аналитики и продукт двигается в слепую. 

Требуется:
1. Описать какие метрики необходимы для продукта
2. Описать как помочь продукту понять, какие пользователи в продукте есть и какие направления в продукте дальше развивать

### **Основные метрики продукта (Product Metrics)**

#### Активность пользователей:
- **DAU/MAU** (Daily/Monthly Active Users)
- **Retention Rate** (1-day, 7-day, 30-day)
- **Session Duration** и количество сессий в день
- **Частота открытия приложений** по платформам

#### Использование функций:
- **Количество отправленных/полученных писем** в день/месяц
- **Открытие писем** (open rate)
- **Клики по ссылкам** в письмах
- **Использование мобильного vs desktop**
- **Доля пользователей с включенными push-уведомлениями**

#### Рекламные метрики:
- **CTR рекламы** по баннерам и типам
- **Конверсия в Premium** (отказ от рекламы)
- **ARPU** (Average Revenue Per User)
- **LTV** пользователей с разным поведением

###  **Мonetization Metrics**

#### Freemium модель:
- **Conversion Rate** в платную подписку
- **Churn Rate** платных пользователей
- **Revenue per user**
- **Количество показов рекламы** и доход от рекламы

###  **Platform Metrics**

#### По платформам:
- **Установки/посещения** по каналам
- **Retention** по платформам
- **Feature adoption** по типам клиентов
- **Технические метрики**: время загрузки, ошибки

###  **Сегментация пользователей**

#### По поведению:
1. **Активные пользователи** (ежедневное использование)
2. **Пассивные пользователи** (раз в неделю/месяц)
3. **Power users** (сотни писем в день)
4. **Light users** (<10 писем в день)
5. **Только получатели** (не отправляют письма)

#### По платформе:
- **Mobile-first** пользователи
- **Desktop-first** пользователи
- **Multi-platform** пользователи
- **Только сторонние клиенты**

#### По монетизации:
- **Free users with ads**
- **Premium users**
- **Users who tried but returned to free**

#### По географии и демографии:
- **Географические регионы**
- **Business vs Personal** использование

###  **Направления развития**

####  **Оптимизация retention**
- Определить drop-off points
- Улучшить onboarding процесс
- Персонализировать опыт

####  **Увеличение монетизации**
- Анализ поведения платных vs бесплатных пользователей
- Оптимизация рекламных форматов
- Таргетинг рекламы

####  **Улучшение UX**
- Heatmaps использования интерфейса
- A/B тестирование новых фич
- Снижение технических проблем

####  **Развитие платформ**
- Сравнение engagement по платформам
- Инвестиции в наиболее перспективные каналы

###  Аналитические дашборды

1. **Real-time мониторинг** ключевых метрик
2. **Cohort analysis** для retention
3. **Funnel analysis** для ключевых процессов
4. **A/B testing framework**


###  **Ключевые метрики успеха для отслеживания**

#### Short-term (1-3 месяца):
- Увеличение DAU на 15%
- Снижение churn rate на 10%
- Увеличение CTR рекламы на 20%

#### Medium-term (3-6 месяцев):
- Рост конверсии в Premium на 25%
- Улучшение 30-day retention на 15%
- Увеличение ARPU на 20%

#### Long-term (6+ месяцев):
- Увеличение LTV на 30%
- Расширение user base на 50%
- Оптимизация product-market fit
