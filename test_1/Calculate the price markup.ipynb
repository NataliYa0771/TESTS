### "Рассчитайте наценку на ювелирные изделия по данным на листе ""Задание 2"".
Соблюсти следующие условия:

1. Период анализа - с 1.04.23 по 10.04.23
2. В анализе должны участвовать только продажи изделий с общей суммой и себестоимостью больше 0 рублей
3. В ответе необходимо получить два значения наценки - среднюю наценку на ювелирные изделия из золота и среднюю наценку на ювелирные изделия из серебра. Сгруппировать изделия по металлу нужно на основе поля ""Товарная группа"". Товарная группа изделий из серебра начинается на ""СИ"", все остальные изделия - это золото."

```python
df = pd.read_excel('Тестовое.xlsx', sheet_name='Задание 2')
```

```python
df
```

```python
april = df[(df['ДатаПродажи'] >= pd.Timestamp('2023-04-01')) & 
           (df['ДатаПродажи'] <= pd.Timestamp('2023-04-10'))].copy()
```

```python
april.shape
```

```python
april = df[df['ДатаПродажи'].between('2023-04-01', '2023-04-10')].copy()
```

```python
april.shape
```

```python
april['ДатаПродажи'] = pd.to_datetime(april['ДатаПродажи']).dt.date
```

```python
april
```

```python
# Посчитаем количество нулей в каждом из столбцов
zero_counts = (april[['ОбщаяСумма', 'Себестоимость']] == 0).sum()

# Выведем результат
print(zero_counts)
```

```python
# Удаляем строки, где 'ОбщаяСумма' == 0
april = april[april['ОбщаяСумма'] != 0].copy()
```

```python
# Посчитаем количество нулей в каждом из столбцов
zero_counts = (april[['ОбщаяСумма', 'Себестоимость']] == 0).sum()

# Выведем результат
print(zero_counts)
```

```python
april.loc[:, 'Металл'] = april['ТоварнаяГруппа'].str.startswith('СИ')
april.loc[:, 'Металл'] = april['Металл'].map({True: 'Серебро', False: 'Золото'})
```

```python
april['Наценка, %'] = (april['ОбщаяСумма'] - april['Себестоимость']) / april['Себестоимость'] * 100
```

```python
april
```

```python
result = april.groupby('Металл')['Наценка, %'].mean().round(2)

print(result)
```
