### ""Рассчитайте прирост средней коверсии на 1 магазин в Санкт- Петербурге из входящих в покупатели""

В анализе прироста средней конверсии Периода2 к Периоду1 должны участвовать только сопоставимые магазины (LFL)-работавшие  в обоих периодах

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Тестовое.xlsx', sheet_name='Задание 1')
```

```python
df
```

```python
# Фильтрация с использованием регулярного выражения
spb_mask = df['Город'].str.contains(r'с.?анкт.?петербург|спб|питер', case=False, regex=True)

# Применяем фильтр
df_spb = df[spb_mask]

df_spb 
```

```python
# Посчитаем количество нулей в каждом из столбцов
zero_counts = (df_spb[['Входящие, чел', 'Покупатели, чел.']] == 0).sum()

# Выведем результат
print(zero_counts)
```

```python
print(df_spb[['Входящие, чел', 'Покупатели, чел.']].dtypes)
```

```python
# Создаём новый столбец с конверсией
df_spb['Конверсия'] = np.where(df_spb['Входящие, чел'] > 0, df_spb['Покупатели, чел.'] / df_spb['Входящие, чел'], 0)
```

```python
lfl_df = df_spb.groupby('Код магазина').filter(lambda x: set(x['Период']) == {1, 2})
```

```python
lfl_df.shape
```

```python
df_spb.shape
```

```python
lfl_df
```

```python
period1 = lfl_df[lfl_df['Период'] == 1]
period2 = lfl_df[lfl_df['Период'] == 2]

mean_conv_period1 = period1['Конверсия'].mean()
mean_conv_period2 = period2['Конверсия'].mean()

growth = (mean_conv_period2 - mean_conv_period1) / mean_conv_period1 * 100

print(f'Средняя конверсия Период 1: {mean_conv_period1:.4f}')
print(f'Средняя конверсия Период 2: {mean_conv_period2:.4f}')
print(f'Прирост конверсии, %: {growth:.2f}%')
```
